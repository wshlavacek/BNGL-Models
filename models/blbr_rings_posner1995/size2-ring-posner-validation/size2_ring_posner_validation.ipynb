# Size-2 Ring Validation

This notebook validates the BNGL model for a bivalent ligand / bivalent receptor system with a size-2 cyclic dimer.

The three encoded doses are `LT = 150`, `331`, and `450`. Here `LT = 331` is the integer ligand count that maximizes equilibrium size-2-ring crosslinking (`Z2`, equivalently `R2`) in the scaled Posner 1995 model.

It does three things:

1. Runs the BNGL file in this directory to generate dose-specific NFsim XML and single-run GDAT outputs.
2. Recomputes Posner 1995 size-2-ring kinetic and equilibrium theory for the same three ligand doses.
3. Reruns NFsim over many seeds on the BNGL-generated XML files and writes a PNG comparing:
   - Posner equilibrium level as a horizontal line,
   - Posner kinetic time course,
   - NFsim mean time course derived from the BNGL model.

The `simulate_nf` actions in the BNGL file are still single-seed runs for quick inspection. The validation figure uses a multi-seed NFsim ensemble mean.
